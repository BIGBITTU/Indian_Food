```python
plt.figure(figsize=(8, 5))
sns.histplot(df['prep_time'], bins=20, kde=True)
plt.xlabel('Preparation Time (minutes)')
plt.ylabel('Frequency')
plt.title('Distribution of Preparation Time')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
df['diet'].value_counts().plot(kind='bar', color=['skyblue', 'salmon'])
plt.xlabel('Diet')
plt.ylabel('Count')
plt.title('Vegetarian vs. Non-vegetarian Dishes')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(7, 4))
df['flavor_profile'].value_counts().plot(kind='bar', color=['lightgreen', 'orange', 'pink', 'purple'])
plt.xlabel('Flavor Profile')
plt.ylabel('Count')
plt.title('Distribution of Flavor Profiles')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
df['course'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['gold', 'lightcoral', 'lightblue', 'mediumseagreen'])
plt.title('Distribution of Meal Courses')
plt.ylabel('')
plt.show()
```

```python
sns.scatterplot(x='prep_time', y='cook_time', data=df)
plt.xlabel('Preparation Time (minutes)')
plt.ylabel('Cooking Time (minutes)')
plt.title('Preparation Time vs. Cooking Time')
plt.show()
```

```python
numeric_columns = ['prep_time', 'cook_time']

# Compute the correlation matrix
corr_matrix = df[numeric_columns].corr()

# Create a heatmap
plt.figure(figsize=(7, 5))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Heatmap for Numeric Features')
plt.show()
```

```python
avg_prep_time_by_state = df.groupby('state')['prep_time'].mean().sort_values(ascending=False)
print(avg_prep_time_by_state)
```
